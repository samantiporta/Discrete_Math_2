$\textbf{Consider the following experiment:}$

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

## Question 1:  Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$


This proof shows that the expected value (average outcome) of a multinomial distribution is the product of the number of trials (n) and the probability of success for a specific category (p_i).

**Proof:**

1. **Expected Value Definition:**
    * We know the expected value (E[y]) is the sum of each outcome (y) multiplied by its probability (P(y)):

    $$E[y] = \sum y \cdot P(y)$$

2. **Joint Probability Distribution:**
    * The provided joint probability distribution (P(y)) considers the number of trials in each category (y_i) and the individual probabilities (P_i) for each category:

    $$P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$$

3. **Simplifying the Sum:**
    * We can rewrite the expected value as a sum of terms, each considering the product of the number of trials in a category (y_i) and its individual probability (P_i):

    $$E[y] = \sum (y_1 \cdot P_1 + y_2 \cdot P_2 + y_3 \cdot P_3 + \dotsb + y_n \cdot P_n) \cdot \frac{n!}{y_1!y_2!y_3!...y_n!}$$

4. **Extracting a Common Factor:**
    * Each term in the sum has the product (n!) of all trials as a factor. We can rewrite this product to include only the specific category's trials (y_i!):

    $$E[y] = \sum \left( y_1 \cdot \frac{n!}{y_1!} \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} + \dotsb + y_n \cdot \frac{n!}{y_n!} \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} \right)$$

5. **Summing Across Categories:**
    * The summation across all categories essentially adds up the expected outcomes for each category, resulting in n times the sum of individual probabilities (p_i):

    $$E[y] = \sum \left( n \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} \right) = n \cdot \left( P_1^1 + P_2^2 + P_3^3 + \dotsb + P_n^n \right)$$

6. **Leveraging the Property of Probabilities:**
    * Since the sum of probabilities across all categories equals 1 (p_1 + p_2 + ... + p_n = 1), the final step simplifies the expected value to n * 1, which is n * p_i:

    $$E[y] = n \cdot (p_1 + p_2 + p_3 + \dotsb + p_n) = n \cdot 1 = n \cdot p_i$$

**Conclusion:** This approach demonstrates how the expected value in a multinomial distribution reflects the average number of successes for a specific category (i) across n trials, considering the individual probability (p_i) of that category.



$\textbf{Question 2:}$ Given the following data:

\begin{bmatrix}
\begin{array}{c|c}
Age & Proportion \\
18-24 & 0.18 \\
25-34 & 0.23 \\
35-44 & 0.16 \\
45-64 & 0.27 \\
65-100 & 0.16 \\
\end{array}
\end{bmatrix}

If 500 adults are sampled  randomly, find the probability that the sample contains 100 persons between 18 and 24, 200 between, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64.. What is the expected value for to obtain a person in the 65 and above?


This scenario demonstrates how the multinomial distribution helps calculate the probability of obtaining specific sample compositions in a multi-category experiment.

**Context:** We're interested in the probability of getting a specific sample distribution for age groups in a survey of 500 adults.

**Given:**

* Total adults (n) = 500
* Number of age groups (k) = 5
* Proportions (pi) for each age group:
    * p1 (18-24) = 0.18
    * p2 (25-34) = 0.23
    * p3 (35-44) = 0.16 (not explicitly mentioned but assumed for calculation)
    * p4 (45-64) = 0.27
    * p5 (65-100) = 0.16

**Multinomial Distribution:**

The probability of a specific sample distribution (y1, y2, ..., yk) in a multinomial experiment is given by the Probability Mass Function (PMF):

$$P(y) = \frac{n!}{y_1! * y_2! * ... * yk!} * p_1^{y_1} * p_2^{y_2} * ... * p_k^{y_k}$$

- n: Total number of samples (adults)
- k: Number of categories (age groups)
- yi: Number of samples in category i (age group)
- pi: Proportion of adults in category i (age group)

**Specific Sample:**

We want to find the probability of having:

* y1 = 100 (adults aged 18-24)
* y2 = 200 (adults aged 25-34)
* y3 = 0 (adults aged 35-44 - assumed based on context)
* y4 = 200 (adults aged 45-64)
* y5 = 0 (adults aged 65-100 - calculated as total - others)

**Calculating Probability:**

Using the PMF formula and substituting the given values:

$$P(y) = \frac{500!}{(100! * 200! * 0! * 200! * 0!)} * (0.18)^{100} * (0.23)^{200} * (0.16)^0 * (0.27)^{200} * (0.16)^0$$

$$P(y) = \frac{500!}{(100! * 200! * 200!)} * (0.18)^{100} * (0.23)^{200} * (0.27)^{200}$$

**Expected Value (Optional):**

While not directly related to probability, the expected value (E[Y5]) for the number of adults in the 65+ age group can be calculated using:

$$E[Y5] = n * p_5 = 500 * 0.16 = 80$$

This indicates, on average, we might expect 80 adults in the 65+ age group based on the given proportions.

$\textbf{Consider the following experiment:}$

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

$\textbf{Question 1:}$ From the following steps shown above, derive the probability distribution of the experiment.

**Proof:**

To derive the probability distribution of the experiment, the binomial distribution is utilized.

The PMF of a binomial distribution is given by:

$$
P(Y = k) = \binom{n}{k} p^k q^{n-k}
$$

Where:
- \( k \) is the number of successes,
- \( n \) is the total number of trials,
- \( p \) is the probability of success on a single trial,
- \( q = (1 - p) \) is the probability of failure on a single trial,
- $\binom{n}{k}$ is the binomial coefficient.

Substituting the given values, the expression becomes:

$$
P(Y = k) = \binom{n}{k} p^k (1 - p)^{n-k}
$$

This represents the probability distribution of the experiment, where $ k $ can range from 0 to $ n $.




$\textbf{Question 2:}$ Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.

**Proof:**

To determine the expectation of the probability distribution, denoted as $( E[Y] )$, we must calculate the sum of each possible outcome \( y \) multiplied by its corresponding probability \( P(y) \).

For a binomial distribution with parameters \( n \) and \( p \), where \( n \) represents the number of trials and \( p \) signifies the probability of success on a single trial, the expectation \( E[Y] \) is expressed as:

$
E[Y] = \sum_{k=0}^{n} y \cdot P(Y = y)
$

Substituting the expression for \( P(Y = y) \) derived previously:

$
E[Y] = \sum_{k=0}^{n} y \cdot \binom{n}{k} p^k (1 - p)^{n-k}
$

Now, for each term in the summation, \( y \) equals \( k \), as \( y \) denotes the number of successes, which is \( k \), and the probability \( P(Y = y) \) is \( \binom{n}{k} p^k (1 - p)^{n-k} \).

$
\begin{align*}
E[Y] & = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n-k} \\
& = \sum_{k=0}^{n} k \cdot \frac{n!}{k!(n-k)!} p^k (1 - p)^{n-k}
\end{align*}
$

Now, let's rewrite the sum in terms of the binomial expansion:

$
E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n-k}
$

$
E[Y] = np\sum_{k=0}^{n} \binom{n-1}{k-1} p^{k-1} (1 - p)^{n-k}
$

Now, the summation represents a binomial expansion of $ (p + (1 - p))^n = 1^n = 1 )$. Therefore, the sum simplifies to 1.

$
E[Y] = np
$

This demonstrates that the expectation of the binomial distribution is $E[Y] = np$.



$\textbf{Question 3:}$ Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?

**Proof:**

To address this issue, we employ the binomial distribution formula. Let $ p = 0.30 $ represent the probability of recovery for a single individual and $ q = 1 - p = 0.70 $ denote the probability of failure to recover for a single individual. Our goal is to ascertain the probability of at least nine out of ten individuals recovering.

The Probability Mass Function (PMF) of the binomial distribution is given by:

$
P(Y = k) = \binom{n}{k} p^k q^{n-k}
$

where $ n $ stands for the total number of trials (number of individuals), $ k $ denotes the number of successes (number of individuals recovering), and $ \binom{n}{k} $ represents the binomial coefficient.

For this scenario, $ n = 10 $, $ p = 0.30 $, and $ q = 0.70 $. Our objective is to compute $ P(Y \geq 9) $.

Now, let's compute the probability using the formula:

$
\begin{align*}
P(Y \geq 9) & = P(Y = 9) + P(Y = 10) \\
& = \binom{10}{9} p^9 q^{10-9} + \binom{10}{10} p^{10} q^{10-10}\\
& = \binom{10}{9} (0.30)^9 (0.70)^{1} + \binom{10}{10} (0.30)^{10}(1)\\
& = 10 \times (0.30)^9 \times (0.70)^{1} + 1 \times (0.30)^{10} \\
& \approx 0.0001377 + 0.0000059 \\
& \approx 0.000144
\end{align*}
$

Thus, the probability that at least nine out of ten individuals receiving the medication will recover is approximately 0.000144.



## $\textbf{Question 4:}$
Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

**Proof:**

To determine the probability of detecting at least one defective fuse in a sample of 5, given that 5% of the fuses are defective, we utilize the complement rule. This rule asserts that the likelihood of an event happening equals 1 minus the probability of the event not occurring.

Let's designate:
- $ p $ as the likelihood of selecting a defective fuse, which equals 0.05.
- $ q $ as the likelihood of selecting a non-defective fuse, calculated as 1 - 0.05 = 0.95.
- $ n $ as the number of trials (i.e., the number of fuses chosen).
- $ Y $ as the number of successes (i.e., the number of observed defective fuses).

Our objective is to determine $ P(Y \geq 1) $, indicating the probability of observing at least one defective fuse.

Employing the complement rule, we formulate:

$ P(Y \geq 1) = 1 - P(Y = 0) $

Now, the probability of encountering 0 defective fuses out of 5 can be computed using the binomial probability formula:

$ P(Y = k) = \binom{n}{k} \times p^k \times q^{(n-k)} $

where:
- $ n $ represents the number of trials (5).
- $ k $ symbolizes the number of successes (0 in this instance).
- $ p $ signifies the likelihood of success (0.05).
- $ q $ denotes the likelihood of failure (0.95).

Hence,

$ P(Y = 0) = \binom{5}{0} \times 0.05^0 \times 0.95^5 $

$ = 1 \times 1 \times (0.95)^5 $

Next, we can integrate this into the complement rule:

$ P(Y \geq 1) = 1 - (0.95)^5 $

Now, let's calculate this probability:

$ P(Y \geq 1) = 1 - (0.95)^5 \approx 1 - 0.7737809375 \approx 0.2262190625 $

Hence, the probability of identifying at least one defective fuse in a sample of 5 is approximately $0.2262$, equivalent to $22.62\%$.

Consider a probability distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.


$\textbf{Question 1:}$ Find a general formula for the expected value of this distribution.

**Proof:**

To compute the expected value of this distribution, denoted as $E[Y]$, we employ the formula:

$E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)$

Upon substituting the provided probability distribution, we obtain:

$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$

Now, let's simplify this expression:

$
\begin{align*}
E[Y] & = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} \\
& = e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!} \\
& = e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} \\
& = e^{-\lambda} \lambda \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!} \\
& = e^{-\lambda} \lambda \sum_{y=0}^{\infty} \frac{\lambda^y}{y!} \\
& = e^{-\lambda} \lambda e^{\lambda} \quad \text{(Using the Taylor series expansion of } e^x \text{)} \\
& = \lambda
\end{align*}
$

Hence, the general formula for the expected value of this distribution is $E[Y] = \lambda$.



$\textbf{Question 2:}$ A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

**Proof:**

To tackle this problem, we leverage the Poisson distribution, given that the mean density of seedlings per square yard conforms to a Poisson distribution with parameter $\lambda = 5$.

The probability mass function (PMF) of the Poisson distribution is expressed as:

$P(Y = y) = \frac{e^{-\lambda} \lambda^y}{y!}$

where:
- $Y$ represents the number of seedlings in a 1-square-yard region,
- $\lambda$ denotes the mean number of seedlings per square yard,
- $e$ stands for the base of the natural logarithm, approximately equal to 2.71828.

For this scenario, $\lambda = 5$ and $y = 0$, since we seek the probability that none of the regions will contain seedlings.

Now, let's compute the probability using the Poisson distribution formula:

$P(Y = 0) = \frac{e^{-5} 5^0}{0!}$

$P(Y = 0) = e^{-5}$

$P(Y = 0) \approx 0.0067$

Thus, the probability that none of the regions will contain seedlings is approximately 0.0067.
