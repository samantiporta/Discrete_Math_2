# Kruskal's Algorithm:
---

```python
import matplotlib.pyplot as plt
import networkx as nx

# Seed for reproducibility
seed = 55
G = nx.Graph()

plt.figure(figsize=(10, 8))

# Add nodes
nodes = ["A", "B", "C", "D", "E", "F", "G"]
G.add_nodes_from(nodes)

# Add edges with weights
edgelist = [
    ("A", "B", 4), ("A", "C", 3), ("B", "C", 1), 
    ("B", "D", 2), ("C", "D", 4), ("C", "E", 2), 
    ("D", "E", 3), ("E", "F", 6), ("F", "G", 2), 
    ("E", "G", 1)
]
G.add_weighted_edges_from(edgelist)

# Perform Kruskal's algorithm to find the MST
mst = nx.minimum_spanning_tree(G, algorithm='kruskal')

# Get positions for the nodes
pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout

# Draw the original graph with edge labels
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=700, node_color='pink', 
        edge_color='black', width=2, font_size=14)
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Highlight the edges in the MST
mst_edges = list(mst.edges(data=True))
nx.draw_networkx_edges(G, pos, edgelist=mst.edges(), width=4, edge_color='blue')

plt.title("Graph with MST Highlighted (Kruskal's Algorithm)")
plt.show()

# Print the edges in the MST
print("Edges in the MST:")
for u, v, weight in mst_edges:
    print(f"{u} - {v}: {weight['weight']}")
```

# Djitraka's Shortest Path Algorithm:
---

```python
import matplotlib.pyplot as plt
import networkx as nx

# Seed for reproducibility
seed = 55
G = nx.Graph()

plt.figure(figsize=(10, 8))

# Add nodes
nodes = ["A", "B", "C", "D", "E", "F", "G"]
G.add_nodes_from(nodes)

# Add edges with weights
edgelist = [
    ("A", "B", 4), ("A", "C", 3), ("B", "C", 1), 
    ("B", "D", 2), ("C", "D", 4), ("C", "E", 2), 
    ("D", "E", 3), ("E", "F", 6), ("F", "G", 2), 
    ("E", "G", 1)
]
G.add_weighted_edges_from(edgelist)

# Source node for Dijkstra's algorithm
source = "A"

# Compute shortest paths from the source node
length, path = nx.single_source_dijkstra(G, source=source)

# Get positions for the nodes
pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout

# Draw the original graph with edge labels
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=700, node_color='pink', 
        edge_color='black', width=2, font_size=14)
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Highlight the shortest paths from the source node
for target in nodes:
    if target != source:
        sp_edges = [(path[target][i], path[target][i+1]) for i in range(len(path[target]) - 1)]
        nx.draw_networkx_edges(G, pos, edgelist=sp_edges, width=4, edge_color='blue')

plt.title(f"Graph with Shortest Paths from {source} Highlighted (Dijkstra's Algorithm)")
plt.show()

# Print the shortest paths and their lengths
print(f"Shortest paths from {source}:")
for target in nodes:
    print(f"Path to {target}: {path[target]} with length {length[target]}")
```
